```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha_e_min = 3e-5
alpha_e_max = 1e-4

alpha_c_min = 5e-6
alpha_c_max = 2e-5

eta_ce_biggest = 0.46
eta_ce_medium = 0.38
eta_ce_smallest = 0.22
```

```python
alpha_e_max/alpha_c_min
```

```python
distances = np.logspace(1, 4)

electric_favored_ratio = (1-alpha_e_min)/(1-alpha_c_max)
fuel_favored_ratio = (1-alpha_e_max)/(1-alpha_c_min)
```

```python
plt.figure(figsize=(4, 3), dpi=400)
plt.plot(distances, np.exp(distances*np.log(fuel_favored_ratio)), label="Electric Transmission 20x Lossier", color="#c0e087")
plt.plot(distances, np.exp(distances*np.log(electric_favored_ratio)), label="Electric Transmission 1.5x Lossier", color="#e30291")

plt.hlines(eta_ce_medium/eta_ce_biggest, 10, 10000, linestyle=":", color="black")
plt.hlines(eta_ce_smallest/eta_ce_medium, 10, 10000, linestyle=":", color="black")

plt.legend( frameon=False)

plt.xscale('log')

```
