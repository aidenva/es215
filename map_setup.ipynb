```python
import random

import matplotlib.pyplot as plt

# Parameters
N = 10                  # Size of the grid
m = 5                   # Number of random nodes
demand_max = 1000

# Generate random nodes with coordinates (x, y) and weights w
nodes = [(random.randint(0, N-1), random.randint(0, N-1), random.randint(0,N-1)) for _ in range(m)]

# Plot the grid
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-1, N)
ax.set_ylim(-1, N)
ax.set_xticks(range(N))
ax.set_yticks(range(N))
ax.grid(True)

# Plot the nodes as circles
for x, y, w in nodes:
    circle = plt.Circle((x, y), w, color='blue', alpha=0.5)
    ax.add_patch(circle)
    ax.text(x, y, f"({x},{y})", color='black', fontsize=8, ha='center', va='center')

plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
